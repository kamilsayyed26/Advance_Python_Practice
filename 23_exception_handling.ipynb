```python
try:
    num = int(input("Enter number: "))
    result = 10 / num
    print(result)

except ValueError:
    print("Invalid input")

except ZeroDivisionError:
    print("Cannot divide by zero")
```

```python
# Else
try:
    num = int(input('Enter a num'))
except valueError: #if
    print('Invalid input')
else:
    print('you have entered',num)
```

```python
# finally
try:
    open('file.txt') #default uses read mode
except FileNotFoundError:
    print('File not Found')
finally:
    print('Code completed')
    
```

```python
import logging 
logging.basicConfig(level1=logging.DEBUG)

logging.debug('DEBUG logging')
logging.info('Info logging')
logging.warning('Warning logging')
logging.error('Error logging')
logging.critical('Critical logging')
```

```python
import logging 
logging.basicConfig(level=logging.ERROR)

try:
    x = 10/0
except ZeroDivisionError:
    logging.error('Cannot devide by zero')
```

```python
import logging 
logging.basicConfig(level=logging.ERROR)

try:
    x = 10/0
except Exception as e:
    logging.error('Error happened: %s',e)
```

```python
import logging 
logging.basicConfig(filename='app.log',level=logging.INFO)

logging.info('project started')

try:
    x = 10/0
except Exception as e:
    logging.error('Error happened: %s',e)
```
