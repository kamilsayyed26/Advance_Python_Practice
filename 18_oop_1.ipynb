```python
class bank_account():
    def __init__(self,balance):
        self.__balance = balance #Private
```

```python
class bank_account():
    def __init__(self,balance):
        self.__balance = balance #Private
    def deposite(self,amount):
        self.__balance = self.__balance + amount
    def balance_check(self):
        return self.__balance
```

```python
obj3 = bank_account(100)
```

```python
obj3.balance_check()
```

```python
obj3.deposite(50)
```

```python
obj3.balance_check()
```
