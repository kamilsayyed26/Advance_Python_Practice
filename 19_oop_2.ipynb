```python
class banking:
    def __init__ (self,balance):
        self.balance = balance
class hdfc(banking):
    def mutual_funds(self):
        result = self.balance/10
        return result
class axis(hdfc):
    def buy_stocks(self,number_shares):
        if number_shares *5 <= self.balance:
            return f'Total number of shares is now: {number_shares} and avg price is 5 and total value of shares is 5'
        else:
            return 'Insufficient fund'
```

```python
c1 =axis(50)
```

```python
c1.balance
```

```python
c1.buy_stocks(30)
```

```python
c1.mutual_funds()
```

```python
c2 = axis(1000)
```

```python
c2.mutual_funds()
```

```python
c2.buy_stocks(30)
```

# Abstraction

```python
from abc import ABC, abstractmethod
```

```python
class car(ABC):  #special class
    @abstractmethod
    def airbag(self):
        pass
```

```python
class nexon(car):
    def airbag(self):
        return 'Air bags are installed in tata nexon'
        # optional features not compulsary
    def sunroof(self):
        return  'Sunroof is installed in tata nexon'
```

```python
c10= nexon()
```

```python
c10.airbag()
```

```python
c10.sunroof()
```

```python
# if not mension airbag 
class fortuner(car):
    def autopark(self):
        return 'sun'
```

```python
class phone(ABC):
    @abstractmethod
    def call(self):
        pass
```

```python
class iphone(phone):
    def call(self):
        return'calling feature is added in your iphone'
    def camera(self):
        return 'camera is added in your iphone'
```

```python
obj11 = iphone()
```

```python
obj11.call()
```

```python
class samsung(phone):
    def call(self):
        return'call feature is added in your samsung phone'
    def front_camera(self):
        return ' front  camera is added in your samsung phone'
```

```python
s23 = samsung()
```

```python
s23.call()
```
